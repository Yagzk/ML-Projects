```python
print(penguins.isnull().sum())
```

```python
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
penguins['bill_depth_mm'] = penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean())
penguins['body_mass_g'] = penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean())
penguins['bill_length_mm'] = penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean())
penguins['flipper_length_mm'] = penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].median())
```

```python
penguins['sex'] = penguins['sex'].map({'male': 0, 'female': 1})
```

```python
features = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'island', 'sex', 'year']]
label = penguins['species']



```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)

print("Eğitim ve Test Setinin Boyutları:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

x --> features
y --> label

**X_train:**

Modelin eğitilmesi (fit edilmesi) için kullanılan giriş verilerini (özellikleri) içerir.

**X_test:**

Model eğitildikten sonra, modelin performansını değerlendirmek için kullanılan giriş verilerini içerir.

**y_train:**

Modelin öğrenmesi gereken hedef değişkenin değerlerini içerir.

**y_test:**

Model eğitildikten sonra, modelin tahminlerini gerçek değerlerle karşılaştırmak için kullanılan hedef değişkenin değerlerini içerir.
