Teorik Açıklamalar:

**Sınıflandırma Metrikleri:** Sınıflandırma problemlerinde model performansını değerlendirmek için kullanılan metrikler, modelin ne kadar doğru tahmin yaptığını ve tahminlerinde ne kadar tutarlı olduğunu anlamaya yardımcı olur.

* **ACCURACY**


Accuracy yönteminde başarı ölçüsü, Doğru Sınıflandırma Sayısının, Toplam Sınıflandırılan Gözlem Sayısına bölümü olarak ifade edilir. Accuracy değeri yükseldikçe modelin başarısı yükselir. Dengeli veri setlerinde işe yarar. Genelde tek başına yeterli olmuyor bu ama.




 Örnek olarak ürün kalite kontrolü verilebilir. Sağlam ve hatalı ürünlerin oranının %50 olduğu bir fabrikada sağlam ve hatalı ürünler dengeli şekilde üretiliyor. Model, ürünleri doğru sınıflandırıyor mu diye test ediliyor.




---



Accuracy = (TP + TN)/(TP + TN + FP + FN)

* Hassasiyet (Precision)

Spam E-posta Tespiti, gibi yanlış pozitiflerin pahalıya mal olduğu durumlarda önemli. (fp durumlarından kaçınıyoruz. Yanlışlıkla önemli bir e postanın spam olarak işaretlenmesini istemeyiz veya gerçek bir kredi kartı işleminin dolandırıcı olarak algılanmasını istemeyiz.)


---



Precision = TP / (TP + FP)


* Duyarlılık (Recall)

kanser testi gibi Kaçırılmaması gereken pozitif sınıflar olduğunda kullanılır. ( yanlış negatifleri(fn) önlüyoruz ) (hasta olan bir kişinin hasta değil olarak sınıflandırılmaması, gerçekten dolandırılmış olup da dolandırılmamış demesini atalamaması)


---



Recall = TP / (TP + FN)


* f1 skoru

Dengesiz veri kümelerinde genel model performansını değerlendirmek için kullanılır.  Precision ve Recall dengesini olduğunda kullanıyoruz.
örnek olarak makine arızaları tahminlerinde ve sosyal medya içerik moderasyonunda kullanılıyor.(makinedeki yanlış arıza alarmlarını azaltmak hem de arızaları kaçırmamak gerekir.)



---



F1 = 2 * (Precision * Recall) / (Precision + Recall)


* Confusion Matrix (Karışıklık Matrisi)


Confusion Matrix, sınıflandırma problemlerinin temel değerlendirme aracı olup, modelin sınıf bazında doğru ve yanlış tahminlerini görselleştirir.

tp --> pozitif sınıfın doğru tahmini ( Hasta olan bir kişi testte pozitif çıkıyor. Doğru bir alarm!)

tn --> negatif sınıfın doğru tahmini (Sağlıklı kişi testte negatif çıkıyor. Sorun yok!)

fp --> negatif sınıfın pozitif olarak yanlış tahmini type 1 error (Sağlıklı birine yanlışlıkla hasta deniyor.)

fn --> pozitif sınıfın negatif olarak yanlış tahmini type 2 error (Gerçek hasta sağlıklısın denip eve gönderiliyor.)

**1. Ortalama Mutlak Hata (Mean Absolute Error - MAE)**

Tahmin edilen değerlerin, gerçek değerlerden ortalama olarak ne kadar sapma gösterdiğini ölçer. Güzel yanı hataları mutlak olarak ölçer, kolay yorumlanabilir. ama kötü yanı büyük hatalar ve küçük hatalar eşit derecede kabul edilir. Değer ne kadar düşükse o kadar iyi performans elde edilir.

**Örnek:**

* Bir süpermarketin tahmin modeli, günlük satılacak süt miktarını tahmin ediyor. MAE, modelin günlük tahminlerinin ortalama ne kadar saptığını gösterir. Örneğin: "Model, günde ortalama 12 litre hata yapıyor."





---


$( MAE = \frac{1}{n} \sum |y_{gerçek} - y_{tahmin}| $\)

2. **Ortalama Kare Hata (Mean Squared Error - MSE)**

Ortalama kare hata veri setindeki gerçek değerler ile tahmin edilen değerleri arasındaki farkların karelerinin toplamı alınarak sonucun örnek sayıya bölünmesidir. Aykırı değerler var ise MSE yüksek çıkabilir. Yani tahmin edilenin gerçekten ne kadar saprığını ölçer. Büyük hatalara karşı hassastır ancak karesini aldığımız için yorumlamak zor olabilir.

**Örneğin**

* Bir hava durumu tahmin modeli, günlük sıcaklığı tahmin ediyor. MSE yüksekse, bazı günlerde modelin çok fazla yanıldığını anlayabiliriz. Bu, "model bazı günlerde büyük hata yapıyor" yorumuna yol açar.


---

$( MAE = \frac{1}{n} \sum |y_{gerçek} - y_{tahmin}| $\)

**3. Kök Ortalama Kare Hata (Root Mean Squared Error - RMSE)**

Ortalama kare hata, değerinin karekökünün alınması ile hesaplanır. MSE’de olduğu gibi aykırı değerlere duyarlıdır. Hata biriminde sonuç verir bu yüzden daha anlamlıdır ve bu yüzden daha kolay yorumlanabilir.

**4. R-kare (R²)**



Modelin tahmin ettiği değerlerin, gerçek değerlerin varyasyonunu ne kadar açıkladığını ölçer. 1 e eşit olması modelin mükemmel öğrendiğini gösterir. 0'a eşit olması hiçbir şey öğrenemediğini gösterir.

Örneğin
* Bir banka, müşterinin kredi puanını tahmin eden bir model geliştiriyor. R² = 0.92 ise model, müşteri verilerinin %92'sini başarıyla açıklayabiliyor demektir.

(y^): Gerçek değerlerin ortalaması.

990/1000 = %99 doğruluk oranı. Sağlıklı kişi sayısı çok olduğu için çok çıktı. Hasta olan 10 bireyi de sağlıklısın diyip eve gönderdik. Çook tehlikeli

**Sınıflandırmada Precision mı yoksa Recall mı daha önemli?**

probleme göre değişkenlik gösterir.

** Örnek verecek olursak:**

Precision (Kesinlik):
* Kesinlik değeri, özellikle False Positive (yanlış pozitif) tahminlerin maliyetinin yüksek olduğu durumlarda kritik önem taşır.
Örneğin, e-posta sınıflandırma sistemlerinde modeliniz gelen kutunuza düşmesi gereken bir e-postayı yanlışlıkla spam (FP) olarak işaretlerse, bu durumda önemli bir mesajı kaçırabilirsiniz. Bu tür sistemlerde Precision değerinin yüksek olması, doğru model seçiminde önemli bir kriterdir.

Recall (Duyarlılık):
* Duyarlılık değeri ise, False Negative (yanlış negatif) tahminlerin maliyetinin yüksek olduğu senaryolarda ön plana çıkar.
Örneğin, kanser taraması yapan bir modelde, hastalığı olan bir bireyin sağlıklı olarak sınıflandırılması (FN) çok ciddi sonuçlara yol açabilir. Bu nedenle, Recall değerinin mümkün olduğunca yüksek olması gerekir.



---


**Mean Absolute Error (MAE) mı yoksa Root Mean Squared Error (RMSE) mi daha önemli?**

duruma göre değişir,

MAE;

* Modelin genelde ortalama olarak ne kadar hata yaptığını görmek istiyorsak, aykırı değerler çoksa ve hatalar eşit önemdeyse tercih edilir.

* Örneğin:
Bir kargo şirketi, teslimat sürelerini tahmin ediyor. 1-2 günlük gecikmeler normal kabul ediliyor, nadir olan 7-8 günlük gecikmeler tüm sistemi bozmamalı. Burada MAE daha anlamlı olur.

RMSE:

* Büyük hatalar çok ciddi sonuçlara yol açıyorsa, modelin küçük hataları tolere etmesi ama büyük hataları mutlaka önlemesini istiyorsak tercih edilir.

* Örneğin: Bir hastanede bir model, ilaç dozajını tahmin ediyor. Küçük hatalar tolere edilebilir ama büyük hata hastaya zarar verebilir. Burada RMSE daha önemlidir çünkü büyük hataları daha çok cezalandırır



---





**Sınıflandırma metriklerinde : bir spam e-posta tespit modelinde F1 skoru ile doğruluk(accuracy) arasındaki fark:**

Örneğin: 1000 e postamız var ve bunun 950 tanesi normal(TN)(gerçekten spam olsa TP derdik), 50 tanesi spam ve gözden kaçıyor (FN), accuracy miz 950/1000 = %95, F1 skorumuz: 0

f1 skorumuz burada modelin gerçekten spam yakalama becerisini ölçer. yani burada spam bulamıyor diyor. Burada accuracy den daha gerçekçi ve sert.

KISACA:
* Accuracy, modelin tüm tahminleri içindeki doğru tahmin oranını verir. Ancak, spam tespiti gibi dengesiz veri setlerinde yanıltıcı olabilir.

* F1 skoru ise, modelin spamleri ne kadar doğru ve ne kadar eksiksiz bulduğunu ölçer.


---



**Regresyon Metriklerinden :Ev fiyatı tahmininde RMSE mi, MAE mi tercih edilmeli?**

Ev fiyatı tahmini gibi durumlarda genellikle RMSE (Root Mean Squared Error) tercih edilir.

* RMSE karesini sonra ortalama ve karekökünü aldığı için büyük hatalara daha fazla ceza verir. Ev fiyatlarında yüksek sapmalar genellikle önemlidir. Örneğin 100.000 TL hata ile 10.000 TL hata aynı değildir; RMSE bu farkı daha iyi yansıtır.

**Hata toleransının düşük olduğu finansal tahminlerde hangi metrik daha anlamlıdır?**

Finansal tahminlerde hata toleransı genellikle düşüktür. Yani küçük sapmalar bile kritik olabilir.
Eğer büyük hatalar özellikle risk teşkil ediyorsa (örneğin kredi riski, portföy değerlemesi), RMSE tercih edilir.
ama tüm hatalar eşit şekilde önemliyse, yani sistem aykırı değerlere duyarlı değilse, MAE de düşünülebilir. Ancak genellikle RMSE tercih ediliyor.
